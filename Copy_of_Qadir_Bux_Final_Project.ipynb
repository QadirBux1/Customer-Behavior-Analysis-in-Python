Exploratory Data Analysis

```python
# figure size
plt.figure(figsize=(10, 6))

# histogram with a kernel density estimate (KDE)
sns.histplot(df['Income'], bins=30, kde=True)

# title of the plot
plt.title('Distribution of Income')

#the plot
plt.show()
```

```python
# figure size
plt.figure(figsize=(16, 10))

# subplot (1st subplot in a 3x2 grid)
plt.subplot(2, 2, 2)

# bar plot for education distribution
sns.barplot(x=df['Education'].value_counts().index, y=df['Education'].value_counts(), palette='viridis')

# plot title and labels
plt.title('Education Distribution')
plt.xlabel('Education Level')
plt.ylabel('Customers')

# the plot
plt.show()
```

```python
# Set the figure size
plt.figure(figsize=(16, 10))

# subplot (1st subplot in a 3x2 grid)
plt.subplot(2, 2, 2)

# Create a count plot for marital status distribution
sns.countplot(x='Marital_Status', data=df, palette='magma')

# Set the title, xlabel, and ylabel
plt.title('Marital Status Distribution')
plt.xlabel('Marital Status')
plt.ylabel('Customers')

# Show the plot
plt.show()
```

```python
# columns for campaign response rates
response_rates = df[['AcceptedCmp1', 'AcceptedCmp2', 'AcceptedCmp3', 'AcceptedCmp4', 'AcceptedCmp5', 'Response']].mean()

#  complaint rate
complaint_rate = df['Complain'].mean()

print("Campaign Response Rates:")
print(response_rates)
print(f"\nComplaint Rate: {complaint_rate}")
```

```python
# columns for campaign acceptance rates
campaign_columns = ['AcceptedCmp1', 'AcceptedCmp2', 'AcceptedCmp3', 'AcceptedCmp4', 'AcceptedCmp5']
campaign_acceptance = df[campaign_columns].mean() * 100

# the bar plot
plt.figure(figsize=(10, 6))
sns.barplot(x=campaign_acceptance.index, y=campaign_acceptance, palette='RdYlBu')

#  title and labels
plt.title('Campaign Acceptance Rates')
plt.xlabel('Campaigns')
plt.ylabel('Acceptance Rate (%)')

# Setting y-axis limit to ensure proper scale (0 to 100%)
plt.ylim(0, 100)

# Show the plot
plt.show()

```

```python
#  Year_Birth in Year format
df['Year_Birth'] = pd.to_datetime(df['Year_Birth'], format='%Y')

#  current year
current_year = pd.to_datetime('today').year

# age from year birth and current year
df['Age'] = current_year - df['Year_Birth'].dt.year
```

```python
print(df['Age'])
```

```python

# Frequency column from all the purchases columns
df['Frequency'] = df[['NumDealsPurchases', 'NumWebPurchases', 'NumCatalogPurchases', 'NumStorePurchases']].sum(axis=1)

# Monetary column from all the amount columns
df['Monetary'] = df[['MntBeverage', 'MntFruits', 'MntMeatProducts', 'MntFishProducts', 'MntSweetProducts', 'MntGoldProds']].sum(axis=1)

print(df[['Frequency', 'Monetary']])
```

```python
selected_features = ['Recency', 'Frequency', 'Monetary', 'Income', 'Age']

#  correlation matrix
correlation_matrix = df[selected_features].corr()

# the heatmap
plt.figure(figsize=(10, 8))
sns.heatmap(correlation_matrix, annot=True, cmap='coolwarm')

# the title
plt.title('Correlation Matrix for Selected Features')

# the plot
plt.show()

```

```python
# columns for correlation matrix
selected_columns = ['Age', 'Income', 'NumWebPurchases', 'NumCatalogPurchases', 'NumStorePurchases']

# the correlation matrix
correlation_matrix = df[selected_columns].corr()

print(correlation_matrix)
```

```python
# columns for correlation matrix
selected_columns = ['Age', 'Income', 'NumWebPurchases', 'NumCatalogPurchases', 'NumStorePurchases']

# the correlation matrix
correlation_matrix = df[selected_columns].corr()

# the heatmap
plt.figure(figsize=(10, 8))
sns.heatmap(correlation_matrix, annot=True, cmap='coolwarm', linewidths=.5)

# the title
plt.title('Correlation Heatmap for Purchasing Variables')

# the plot
plt.show()
```

```python
# age bins and labels
age_bins = [18, 30, 40, 50, 60, 70, 80, 90, float('inf')]
age_labels = ['18-29', '30-39', '40-49', '50-59', '60-69', '70-79', '80-89', '90+']

# new column 'AgeGroup' based on age bins
df['AgeGroup'] = pd.cut(df['Age'], bins=age_bins, labels=age_labels)
```

```python
#  number of customers in each age group
age_group_counts = df['AgeGroup'].value_counts().reset_index().rename(columns={'index': 'AgeGroup', 'AgeGroup': 'Count'})

# the bar plot
plt.figure(figsize=(10, 6))
sns.barplot(x='AgeGroup', y='Count', data=age_group_counts, color='skyblue')

# the title and labels
plt.title('Number of Customers in Each Age Group')
plt.xlabel('Age Group')
plt.ylabel('Number of Customers')

# the plot
plt.show()
```

```python
product_columns = ['MntBeverage', 'MntFruits', 'MntMeatProducts', 'MntFishProducts', 'MntSweetProducts', 'MntGoldProds']
df['TotalSpending'] = df[product_columns].sum(axis=1)
```

```python
# income groups
income_groups = pd.cut(df['Income'], bins=[0, 40000, 80000, 120000, float('inf')], labels=['0-40k', '40k-80k', '80k-120k', '120k+'])
df['IncomeGroup'] = income_groups

# average total spending by income group
average_spending_by_income = df.groupby('IncomeGroup')['TotalSpending'].mean().reset_index()

# Plotting the bar plot
plt.figure(figsize=(10, 6))
sns.barplot(x='IncomeGroup', y='TotalSpending', data=average_spending_by_income)

# plot title and labels
plt.title('Average Total Spending by Income Group')
plt.xlabel('Income Group')
plt.ylabel('Average Total Spending')

# Show the plot
plt.show()
```

```python
# Group by AgeGroup and sum of income
total_income_by_age = df.groupby('AgeGroup')['Income'].sum().reset_index()

# Plot a grouped bar plot
plt.figure(figsize=(12, 8))
sns.barplot(x='AgeGroup', y='Income', data=total_income_by_age, color='blue', label='Total Income')

# the title and labels
plt.title('Total Income by Age Group')
plt.xlabel('Age Group')
plt.ylabel('Total Income')

# the legend
plt.legend()

# the plot
plt.show()
```

```python
# the bar plot
plt.figure(figsize=(12, 8))
sns.barplot(x='AgeGroup', y='TotalSpending', data=df, errorbar=None, palette='viridis')

# the title and labels
plt.title('Average Total Spending by Age Group')
plt.xlabel('Age Group')
plt.ylabel('Average Total Spending')

# the plot
plt.show()
```

```python
# New column Total Purchases by adding all the purchases
df['TotalPurchases'] = df[['NumDealsPurchases', 'NumWebPurchases', 'NumCatalogPurchases', 'NumStorePurchases']].sum(axis=1)

print(df['TotalPurchases'])
```

```python
# Group by AgeGroup and mean of each type of purcahse
average_purchases = df.groupby('AgeGroup')[['NumWebPurchases', 'NumCatalogPurchases', 'NumStorePurchases', 'NumDealsPurchases']].mean().reset_index()

# Plot a grouped bar plot
plt.figure(figsize=(12, 8))
sns.barplot(x='AgeGroup', y='NumWebPurchases', data=average_purchases, color='blue', label='Web Purchases')
sns.barplot(x='AgeGroup', y='NumCatalogPurchases', data=average_purchases, color='green', label='Catalog Purchases', bottom=average_purchases['NumWebPurchases'])
sns.barplot(x='AgeGroup', y='NumStorePurchases', data=average_purchases, color='orange', label='Store Purchases', bottom=average_purchases[['NumWebPurchases', 'NumCatalogPurchases']].sum(axis=1))
sns.barplot(x='AgeGroup', y='NumDealsPurchases', data=average_purchases, color='purple', label='Deals Purchases', bottom=average_purchases[['NumWebPurchases', 'NumCatalogPurchases', 'NumStorePurchases']].sum(axis=1))

# the title and labels
plt.title('Average Number of Purchases by Age Group')
plt.xlabel('Age Group')
plt.ylabel('Average Number of Purchases')

# the legend
plt.legend()

# the plot
plt.show()
```

```python
# the number of customers in each country
customer_count_by_country = df['Country'].value_counts().reset_index()
customer_count_by_country.columns = ['Country', 'CustomerCount']

# the bar plot
plt.figure(figsize=(12, 8))
sns.barplot(x='Country', y='CustomerCount', data=customer_count_by_country)

# the title and labels
plt.title('Number of Customers by Country')
plt.xlabel('Country')
plt.ylabel('Number of Customers')

# the plot
plt.show()
```

```python
# Total income by country
total_income_by_country = df.groupby('Country')['Income'].sum().reset_index()

custom_palette = sns.color_palette("viridis", len(total_income_by_country))

# The figure
plt.figure(figsize=(12, 8))
sns.barplot(x='Country', y='Income', data=total_income_by_country, palette=custom_palette)

# The title and labels
plt.title('Total Income by Country')
plt.xlabel('Country')
plt.ylabel('Total Income')

# The plot
plt.show()
```

```python
# Average of income
average_income_by_country = df.groupby('Country')['Income'].mean().reset_index()

custom_palette = sns.color_palette("viridis", len(average_income_by_country))

# The figure
plt.figure(figsize=(12, 8))
sns.barplot(x='Country', y='Income', data=average_income_by_country, palette=custom_palette)

# The title and labels
plt.title('Average Income by Country')
plt.xlabel('Country')
plt.ylabel('Average Income')

# The plot
plt.show()
```

```python
# The figure size
plt.figure(figsize=(12, 8))

# Grouped by Country
total_purchases_by_country = df.groupby('Country')[['NumWebPurchases', 'NumCatalogPurchases', 'NumStorePurchases', 'NumDealsPurchases']].sum().reset_index()

bar_width = 0.5
positions = np.arange(len(total_purchases_by_country))

# Plot each type of purchase separately
plt.bar(positions, total_purchases_by_country['NumWebPurchases'], label='Web Purchases', width=bar_width)
plt.bar(positions, total_purchases_by_country['NumCatalogPurchases'], label='Catalog Purchases', bottom=total_purchases_by_country['NumWebPurchases'], width=bar_width)
plt.bar(positions, total_purchases_by_country['NumStorePurchases'], label='Store Purchases', bottom=total_purchases_by_country['NumWebPurchases'] + total_purchases_by_country['NumCatalogPurchases'], width=bar_width)
plt.bar(positions, total_purchases_by_country['NumDealsPurchases'], label='Deals Purchases', bottom=total_purchases_by_country['NumWebPurchases'] + total_purchases_by_country['NumCatalogPurchases'] + total_purchases_by_country['NumStorePurchases'], width=bar_width)

# The title and labels
plt.title('Total Number of Purchases by Country')
plt.xlabel('Country')
plt.ylabel('Total Number of Purchases')

# The legend
plt.legend()
plt.xticks(positions, total_purchases_by_country['Country'])

# Show the plot
plt.show()
```

```python
# The figure size
plt.figure(figsize=(12, 8))

# Grouped by Country
average_purchases_by_country = df.groupby('Country')[['NumWebPurchases', 'NumCatalogPurchases', 'NumStorePurchases', 'NumDealsPurchases']].mean().reset_index()

bar_width = 0.5
positions = np.arange(len(average_purchases_by_country))

# Plot each type of purchase separately
plt.bar(positions, average_purchases_by_country['NumWebPurchases'], label='Web Purchases', width=bar_width)
plt.bar(positions, average_purchases_by_country['NumCatalogPurchases'], label='Catalog Purchases', bottom=average_purchases_by_country['NumWebPurchases'], width=bar_width)
plt.bar(positions, average_purchases_by_country['NumStorePurchases'], label='Store Purchases', bottom=average_purchases_by_country['NumWebPurchases'] + average_purchases_by_country['NumCatalogPurchases'], width=bar_width)
plt.bar(positions, average_purchases_by_country['NumDealsPurchases'], label='Deals Purchases', bottom=average_purchases_by_country['NumWebPurchases'] + average_purchases_by_country['NumCatalogPurchases'] + average_purchases_by_country['NumStorePurchases'], width=bar_width)

# The title and labels
plt.title('Average Number of Purchases by Country')
plt.xlabel('Country')
plt.ylabel('Average Number of Purchases')

# The legend
plt.legend()
plt.xticks(positions, average_purchases_by_country['Country'])

# The plot
plt.show()
```

```python
total_spending = df.groupby('Country')[['MntBeverage', 'MntFruits', 'MntMeatProducts', 'MntFishProducts', 'MntSweetProducts', 'MntGoldProds']].sum().reset_index()
total_spending['TotalSpending'] = total_spending[['MntBeverage', 'MntFruits', 'MntMeatProducts', 'MntFishProducts', 'MntSweetProducts', 'MntGoldProds']].sum(axis=1)

# Sort the DF by TotalSpending in descending order
total_spending = total_spending.sort_values(by='TotalSpending', ascending=False)

# The fiure
plt.figure(figsize=(12, 8))
sns.barplot(x='Country', y='TotalSpending', data=total_spending)

# the title and labels
plt.title('Total Spending by Country')
plt.xlabel('Country')
plt.ylabel('Total Spending')

# The plot
plt.show()
```

```python
# total spending for each product category and overall total spending
spending_by_country = df.groupby('Country')[['MntBeverage', 'MntFruits', 'MntMeatProducts', 'MntFishProducts', 'MntSweetProducts', 'MntGoldProds']].sum()
spending_by_country['TotalSpending'] = spending_by_country.sum(axis=1)

# Avverage spending per customer for each country
spending_by_country['AverageSpending'] = spending_by_country['TotalSpending'] / df.groupby('Country').size()

# The figure
plt.figure(figsize=(12, 8))
spending_by_country['AverageSpending'].plot(kind='bar')

# The title and labels
plt.title('Average Spending by Country')
plt.xlabel('Country')
plt.ylabel('Average Spending')

# The plot
plt.show()
```
